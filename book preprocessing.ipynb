```python
import pdfplumber

def extract_text_from_pdf(pdf_path):
    with pdfplumber.open(pdf_path) as pdf:
        text = ""
        for page in pdf.pages:
            text += page.extract_text()
    return text
```

```python
import pdfplumber

def extract_text_from_pdf(pdf_path, start_page, end_page):
    with pdfplumber.open(pdf_path) as pdf:
        text = ""
        for page in pdf.pages[start_page:end_page+1]:
            text += page.extract_text()
    return text

# PDF 파일 경로 지정
pdf_file_path = "C:/Users/hiyou/Downloads/christie.pdf"

# 추출할 시작 페이지와 종료 페이지 지정 (0부터 시작)
start_page = 6
end_page = 40

# 텍스트 추출
text = extract_text_from_pdf(pdf_file_path, start_page, end_page)

# 추출된 텍스트 출력
print(text)
```

```python
import re

def preprocess_text(text):
    # 문장 구분을 위한 줄바꿈 기호 처리
    text = text.replace('\n', ' ')  # 줄바꿈 기호를 공백으로 대체
    text = re.sub(r'\s+', ' ', text)  # 연속된 공백을 하나의 공백으로 축소

    # 문장 구분을 위한 마침표 처리
    text = re.sub(r'(?<=[^.?!])\s+(?=[A-Z])', '. ', text)  # 문장 끝의 공백 제거
    text = re.sub(r'(?<=[^.?!])\s*$', '', text)  # 문장 끝의 공백 및 마지막 마침표 제거

    return text

# 텍스트 전처리
preprocessed_text = preprocess_text(text)

# 전처리된 텍스트 출력
print(preprocessed_text)
```

```python
from nltk import sent_tokenize

sentences = sent_tokenize(preprocessed_text)
print(sentences)
```

```python
from tensorflow.keras.preprocessing.text import text_to_word_sequence

words =  [text_to_word_sequence(sentence) for sentence in sentences]
print(words)
```

```python
import rdflib

g = rdflib.Graph()

namespace = rdflib.Namespace("http://example.com/")

subject = namespace["subject"]
g.add((subject, rdflib.RDF.type, namespace["Entity"]))

property = namespace["property"]
g.add((subject, namespace["hasProperty"], property))

relation = namespace["relation"]
g.add((subject, namespace["hasRelation"], relation))

g.serialize("output.rdf", format="xml")
```
